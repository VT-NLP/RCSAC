```python
import re
import os
import pandas as pd
from io import StringIO
from sklearn.model_selection import train_test_split
```

```python
# filename = 'ACAR556.csv'
for filename in os.listdir('./PJ_cleaned/'):
    if '.csv' in filename:
        print(filename)
        df = pd.read_csv(f'./PJ_cleaned/{filename}', sep='|', header=None)

        new_training_string = ""

        # print(df[2])
        # u_idx = 0
        messages = ["START"]
        for i in range(len(df)):
            if df.iloc[i][0] in filename:
                usr = "groomer"
            else:
                usr = "victim"
            
            msg = df.iloc[i][2]
            # usr = users[u_idx]
            if len(messages) > 10:
                hist = ';;'.join(messages[-10:])
            else:
                hist = ';;'.join(messages)

            if usr == 'groomer':
                new_training_string += f"{hist}|{usr}:{msg}\n"
            messages.append(f'{usr}:{msg}')

            # u_idx = not u_idx
#print(new_training_string)


file = open(f'./groomer_training_data.csv', 'w')
file.writelines(new_training_string)
file.close()

df2 = pd.read_csv('./groomer_training_data.csv', sep='|', header=None)
train, test = train_test_split(df2, test_size=0.1)
test.to_csv('./groomer_test.csv', sep='|')
train.to_csv('./groomer_train.csv', sep='|')
```
